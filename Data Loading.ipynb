```python
import json
import urllib.request
url = 'https://d2ad6b4ur7yvpq.cloudfront.net/naturalearth-3.3.0/ne_50m_urban_areas.geojson'
uh = urllib.request.urlopen(url)
result = json.load(uh)


with open('data.geojson', 'w') as json_file:
    json.dump(result, json_file)
```
